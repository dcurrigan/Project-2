```python
# Dependencies
import pandas as pd
import numpy as np
from sqlalchemy import create_engine

import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import inspect

from sqlalchemy import Column, Float, Integer, String
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
```

```python
engine = create_engine('sqlite:///SQL/SGU.db', echo=True)
conn = engine.connect()
```

```python
# Define our installation table
class Installation(Base):
    __tablename__ = 'installation'
    postcode = Column(Integer, primary_key=True)
    _2001 = Column(Float)
    _2002 = Column(Float)
    _2003 = Column(Float)
    _2004 = Column(Float)
    _2005 = Column(Float)
    _2006 = Column(Float)
    _2007 = Column(Float)
    _2008 = Column(Float)
    _2009 = Column(Float)
    _2010 = Column(Float)
    _2011 = Column(Float)
    _2012 = Column(Float)
    _2013 = Column(Float)
    _2014 = Column(Float)
    _2015 = Column(Float)
    _2016 = Column(Float)
    _2017 = Column(Float)
    _2018 = Column(Float)
    _2019 = Column(Float)
    _2020 = Column(Float)
    _2021 = Column(Float)
    AVG = Column(Float)
    Total = Column(Float)
```

```python
# Define our output table
class Output(Base):
    __tablename__ = 'output'
    postcode = Column(Integer, primary_key=True)
    _2001 = Column(Float)
    _2002 = Column(Float)
    _2003 = Column(Float)
    _2004 = Column(Float)
    _2005 = Column(Float)
    _2006 = Column(Float)
    _2007 = Column(Float)
    _2008 = Column(Float)
    _2009 = Column(Float)
    _2010 = Column(Float)
    _2011 = Column(Float)
    _2012 = Column(Float)
    _2013 = Column(Float)
    _2014 = Column(Float)
    _2015 = Column(Float)
    _2016 = Column(Float)
    _2017 = Column(Float)
    _2018 = Column(Float)
    _2019 = Column(Float)
    _2020 = Column(Float)
    _2021 = Column(Float)
    AVG = Column(Float)
    Total = Column(Float)
```

```python
# Import our data into pandas from CSV

installation_data_path = 'static/data/SGU/Installation.csv'
output_data_path = 'static/data/SGU/Output.csv'

Installation_df = pd.read_csv(installation_data_path, low_memory=False)
Output_df = pd.read_csv(output_data_path, low_memory=False)
```

```python
# Create the tables within the database
Base.metadata.create_all(engine)
```

```python
Installation_df.to_sql('installation', engine, if_exists="append", index=False)
```
