```python
# Dependencies
import pandas as pd
import numpy as np
from sqlalchemy import create_engine

import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import inspect

from sqlalchemy import Column, Float, Integer, String
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
```

```python
engine = create_engine('sqlite:///SQL/SGU.db')
conn = engine.connect()
```

```python
# Define our installation table
class Installation(Base):
    __tablename__ = 'installation'
    postcode = Column(Integer, primary_key=True)
    _2001 = Column(Float)
    _2002 = Column(Float)
    _2003 = Column(Float)
    _2004 = Column(Float)
    _2005 = Column(Float)
    _2006 = Column(Float)
    _2007 = Column(Float)
    _2008 = Column(Float)
    _2009 = Column(Float)
    _2010 = Column(Float)
    _2011 = Column(Float)
    _2012 = Column(Float)
    _2013 = Column(Float)
    _2014 = Column(Float)
    _2015 = Column(Float)
    _2016 = Column(Float)
    _2017 = Column(Float)
    _2018 = Column(Float)
    _2019 = Column(Float)
    _2020 = Column(Float)
    _2021 = Column(Float)
    AVG = Column(Float)
    Total = Column(Float)
```

```python
# Define our output table
class Output(Base):
    __tablename__ = 'output'
    postcode = Column(Integer, primary_key=True)
    _2001 = Column(Float)
    _2002 = Column(Float)
    _2003 = Column(Float)
    _2004 = Column(Float)
    _2005 = Column(Float)
    _2006 = Column(Float)
    _2007 = Column(Float)
    _2008 = Column(Float)
    _2009 = Column(Float)
    _2010 = Column(Float)
    _2011 = Column(Float)
    _2012 = Column(Float)
    _2013 = Column(Float)
    _2014 = Column(Float)
    _2015 = Column(Float)
    _2016 = Column(Float)
    _2017 = Column(Float)
    _2018 = Column(Float)
    _2019 = Column(Float)
    _2020 = Column(Float)
    _2021 = Column(Float)
    AVG = Column(Float)
    Total = Column(Float)
```

```python
# Define our rebate table
class Rebate(Base):
    __tablename__ = 'rebate'
    postcode = Column(Integer, primary_key=True)
    zone = Column(Float)
    rating = Column(Float)
    annual_prod = Column(Float)
    rebate = Column(Float)

```

```python
# Define our income table
class Income(Base):
    __tablename__ = 'income'
    Postcode = Column(Integer, primary_key=True)
    Average_total = Column(Float)
    Average_salary = Column(Float)
```

```python
# Create the tables within the database
Base.metadata.create_all(engine)
```

```python
Installation_df.to_sql('installation', engine, if_exists="append", index=False)
Output_df.to_sql('output', engine, if_exists="append", index=False)
income_df.to_sql('income', engine, if_exists="append", index=False)
rebate_df.to_sql('rebate', engine, if_exists="append", index=False)
```

```python
# Declare a Base using `automap_base()`
Base = automap_base()
```

```python
# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)
```

```python
# Print all of the classes mapped to the Base
Base.classes.keys()
```

```python
# Assign the installation class to a variable called `Install`
Install = Base.classes.installation
```

```python
# Create a session
session = Session(engine)
```

```python
# Use the session to query Install table and display the first 5 trade volumes
for row in session.query(Install.postcode, Install.AVG, Install.Total).limit(20).all():
    print(row)
```

```python
# Assign the output class to a variable called `Output`
Output = Base.classes.output
```

```python
# Use the session to query Output table and display the first 5 trade volumes
for row in session.query(Output.postcode, Output.AVG, Output.Total).limit(20).all():
    print(row)
```
