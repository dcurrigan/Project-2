```python
class Rdata(Base):
    __tablename__ = 'rdata'
    postcode = Column(Integer, primary_key=True)
    ins_avg = Column(Float)
    ins_total = Column(Float)
    out_avg = Column(Float)
    out_total = Column(Float)
    suburb = Column(String(255))
    state = Column(String(255))
```

```python
# Create the tables within the database
Base.metadata.create_all(engine)
```

```python
rdata.to_sql('rdata', engine, if_exists="append", index=False)
```

```python
# Reflect the database tables
Base = automap_base()
Base.prepare(engine, reflect=True)
```

```python
# Print all of the classes mapped to the Base
Base.classes.keys()
```

```python
# Assign the output class to a variable called `Output`
Install = Base.classes.installation
```

```python
# Create a session
session = Session(engine)
```

```python
# Use the session to query Suburbs table
for row in session.query(Rdata.postcode, Rdata.ins_avg, Rdata.ins_total, Rdata.out_avg, Rdata.out_total, Rdata.suburb, Rdata.state).limit(20).all():
    print(row)
```
